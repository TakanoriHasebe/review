```python
# DataVisualization
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

# Stats
import scipy as sp
from scipy import stats
import statsmodels.formula.api as smf
import statsmodels.api as sm
# 表示桁数の指定
%precision 3
```

```python
sales = pd.read_csv("5-3-1-lm-model.csv")
sales.head(3)
```

### データの図示

```python
sns.pairplot(data=sales, hue='weather', palette='gray')
```

### 説明変数同士の関係を調べる

```python
sales.groupby('weather').mean()
```

上記から「雨の日には売上が下がるので、その対策としてお店が値引きをした」と判断できる。<br>

```python
sns.lmplot(x='price', y='sales', data=sales, hue='weather')
```

### 複数の説明変数を持つモデル

```python
lm_sales = smf.ols("sales ~ weather + humidity + temperature + price", data=sales).fit()
lm_sales.params
```

### Type II ANOVA

```python
mod_full = smf.ols("sales ~ weather + humidity + temperature + price", sales).fit()
sm.stats.anova_lm(mod_full, typ=2).round(3)
```

p値が0.578であるので、湿度は売上に対して有意な影響を持っていないという結果になる。<br>
